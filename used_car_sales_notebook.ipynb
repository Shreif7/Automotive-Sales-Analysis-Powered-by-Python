```python
A = ["Maruti", "Hyundai", "Toyota","Mahindra","Tata","Renault","Prazo","Kia"]
B = [473, 358, 346, 301,194,181,162,151]   # Sold
P = [1790, 1158, 1203, 1297,636,610,518,622]  # Unsold

xpos = np.arange(len(A))

plt.figure(figsize=(10,6), facecolor="black")
ax = plt.gca()
ax.set_facecolor("black")

plt.title("Comparison Between Sold and Unsold Cars",
          fontsize=15, fontweight='bold', color="white")

bars_sold = plt.bar(xpos - 0.2, B, color="#C1FF72", width=0.4, label="Sold")
bars_unsold = plt.bar(xpos + 0.2, P, color="#a5be00", width=0.4, label="Unsold")

plt.xticks(xpos, A, rotation=45, color="white", fontsize=10, fontweight='bold')
plt.yticks(color="white")
plt.legend(facecolor="black", labelcolor="white")

axes = plt.gca()
axes.spines["top"].set_visible(False)
axes.spines["right"].set_visible(False)

for spine in ax.spines.values():
    spine.set_color("white")

for i in range(len(A)):
    total = B[i] + P[i]
    sold_pct = (B[i] / total) * 100
    unsold_pct = (P[i] / total) * 100

    plt.text(
        xpos[i] - 0.2,
        B[i] / 2,
        f"{sold_pct:.1f}%",
        ha='center', color="black", fontsize=10, fontweight="bold",rotation=45
    )

    plt.text(
        xpos[i] + 0.2,
        P[i] / 2,
        f"{unsold_pct:.1f}%",
        ha='center', color="white", fontsize=10, fontweight="bold" ,rotation=45
    )

plt.tight_layout()
plt.savefig("pnj")
```
